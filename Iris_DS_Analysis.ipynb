#**Sepal Length by species**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming the data is loaded into a DataFrame named 'iris_df'
iris_df = pd.read_csv('/content/iris.data', header=None, names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species'])


plt.figure(figsize=(10, 6))
sns.violinplot(x='species', y='sepal_length', data=iris_df)
plt.title('Distribution of Sepal Length by Species')
plt.xlabel('Species')
plt.ylabel('Sepal Length (cm)')
plt.show()
```

#**Relation Between Petal Length and Petal Width**

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='petal_length', y='petal_width', hue='species', data=iris_df, s=100)
plt.title('Petal Length vs. Petal Width by Species')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.legend(title='Species')
plt.show()
```

#**All features pairwise relation**

```python
sns.pairplot(iris_df, hue='species', diag_kind='kde')
plt.suptitle('Pairwise Relationships of Iris Dataset Features', y=1.02)
plt.show()
```
